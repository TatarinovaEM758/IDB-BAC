```python
db.movies.aggregate([
  {
    $match: {
      year: { $type: "number", $exists: true }
    }
  },
  {
    $group: {
      _id: { $floor: { $divide: ["$year", 10] } },
      count: { $sum: 1 }
    }
  },
  {
    $project: {
      decade: { $concat: [{ $toString: { $multiply: ["$_id", 10] } }, "s"] },
      count: 1,
      _id: 0
    }
  },
  {
    $sort: { decade: 1 }
  }
])
```
